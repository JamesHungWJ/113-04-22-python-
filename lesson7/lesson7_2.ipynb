```python
import pyinputplus as pyip
deposit = 0  #初始儲存金額
num = 0      #初始開始儲存月份
while deposit < 30000:  #目前只有在此才能跳出迴圈，並不靈活，如有1千行程式，就很麻煩。
    input_money = pyip.inputInt(f"請輸入第{num+1}月存的金額:") #如num沒+1。第一個月份為0
    print(input_money)    
    deposit += input_money
    num += 1
print(f"最後一次存的金額是:{input_money}")
print(f"小明總共存了{num}個月,已經存了{deposit}元")
```

- for 或 while迴圈內，遇到「break」，就能立刻跳出迴圈，後面如還有程式，均不執行。
- 

```python
import pyinputplus as pyip
deposit = 0
num = 0
while True:  #True代表，不論如何，每次都進到城市裡面。
    num += 1  #一進程式，立刻+1
    input_money = pyip.inputInt(f"請輸入第{num}月存的金額:")
    print(input_money)    
    deposit += input_money
    if deposit >= 30000:   #還沒到3萬
        break  #立刻斷掉，回到前面
               #如果num+1，放在這裡，會多1個月。
print(f"最後一次存的金額是:{input_money}")
print(f"小明總共存了{num}個月,已經存了{deposit}元")
```

```python
#請輸入整數,輸入[q]離開:
#顯示:
#奇數輸出平方
#偶數不做任何動作
#'q'要離開

while True:
    value = input("請輸入整數,輸入[q]離開:")
    if value == 'q':
        break
    try:
        valueInt = int(value)
    except:
        print("輸入錯誤")
        break
    
    if valueInt % 2 == 1:
        print(f"{valueInt}的平方是{valueInt ** 2}")
    else:
        print("不做事")

print("應用程式結束")
```

## continue
- 意思為，後續的程式不執行，跳到最前面的程式。
- 

```python
#請輸入整數,輸入[q]離開:
#顯示:
#奇數輸出平方
#偶數不做任何動作
#'q'要離開

while True:
    value = input("請輸入整數,輸入[q]離開:")
    if value == 'q':
        break
    try:
        valueInt = int(value)
    except:
        print("輸入錯誤")
        continue
    
    if valueInt % 2 == 1:
        print(f"{valueInt}的平方是{valueInt ** 2}")
    else:
        print("不做事")

print("應用程式結束")
```

- 輸出:
- 10的因數是:
- 1 2 5 10

- 因數與倍數：若兩個整數能整除，則稱除數為被除數的因數；被除數為除數的倍數。【範例】：35÷5＝7…0，5 是 35 的因數，35 是 5 的倍數。 

```python
import pyinputplus as py
num = py.inputInt("請輸入整數,求所有因數:",min=1)
print(num)
print(f"{num}的因數是:")

for i in range(1,num+1):
    if num % i == 0:
        print(i,end=" ")

```
