```python
#小明想要存錢買一輛機車,機車每輛30000元，他將每月存的錢輸入，當存款足夠買機車時，就顯示提示訊息告知。

import pyinputplus as pyip
deposit = 0
while deposit < 30000:
    input_money = pyip.inputInt("請輸入這個月存的金額:")
    print(input_money)
    deposit += input_money
print(f"最後一次存的金額是:{input_money}")
print(f"小明總共存了{deposit}元")
```

```python
import pyinputplus as pyip
deposit = 0  #初始儲存金額
num = 0      #初始開始儲存月份
while deposit < 30000:  #目前只有在此才能跳出迴圈，並不靈活，如有1千行程式，就很麻煩。
    input_money = pyip.inputInt(f"請輸入第{num+1}月存的金額:") #如num沒+1。第一個月份為0
    print(input_money)    
    deposit += input_money
    num += 1
print(f"最後一次存的金額是:{input_money}")
print(f"小明總共存了{num}個月,已經存了{deposit}元")
```

- for 或 while迴圈內，遇到break，就能立刻跳出迴圈，後面如還有程式，均不執行。


```python
import pyinputplus as pyip
deposit = 0
num = 0
while True:  #True代表，不論如何，每次都進到城市裡面。
    num += 1  #一進程式，立刻+1
    input_money = pyip.inputInt(f"請輸入第{num}月存的金額:")
    print(input_money)    
    deposit += input_money
    if deposit >= 30000:   #還沒到3萬
        break  #立刻斷掉，回到前面
               #如果num+1，放在這裡，會多1個月。
print(f"最後一次存的金額是:{input_money}")
print(f"小明總共存了{num}個月,已經存了{deposit}元")
```
