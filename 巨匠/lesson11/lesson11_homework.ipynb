# 使用pydantic解析lesson11資料夾內的data.json檔



```python
from pydantic import BaseModel, Field

class Person(BaseModel):
    name: str
    website: str
    from_: str = Field(alias="from")

class Data(BaseModel):  #字根是data
    people: list[Person]

with open("data.json", encoding="utf-8") as file:
    data: Data = Data.model_validate_json(file.read())

data_list: list[Person] = data.people
for item in data_list:
    print(item)
```

```python
#data實體轉使用model_dump()，可以轉為python的基本資料結構
from pprint import pprint  #前面的pptint是內建的module，後面的pprint是funtion

data_dict:dict = data.model_dump()  
pprint(data_dict) 
```
