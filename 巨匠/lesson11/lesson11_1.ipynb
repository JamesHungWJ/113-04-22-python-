import json： 引入 Python 的 json 模組，該模組提供了處理 JSON 資料的功能，例如解析（從 JSON 格式轉換為 Python 資料結構）和序列化（將 Python 資料結構轉換為 JSON 格式）

with open('aqx_p_488.json', encoding='utf-8') as file:： 使用 with open 語法來打開名為 aqx_p_488.json 的 JSON 檔案，並指定以 utf-8 編碼來讀取。with open 語法會自動在程式執行完畢後關閉檔案，這是推薦的檔案讀取方式，因為它能確保資源管理得當。

data: dict = json.loads(file.read())： 讀取檔案的內容並將其解析為 Python 字典（dict）。json.loads 用來將字串格式的 JSON 資料轉換為 Python 的字典格式，這樣我們就能使用 Python 內建的資料結構來操作這些資料。

data_list: list[dict] = data["records"]： 從字典 data 中提取出鍵（key）為 "records" 的值，並將其賦值給 data_list。data_list 是一個包含多個字典的列表，每個字典對應一組資料。這行程式碼同時也標註了類型提示，表示 data_list 是一個列表，且每個列表項目都是字典。

for site in data_list:： 使用 for 迴圈遍歷 data_list，其中每個 site 是一個字典，包含單一站點的空氣品質資料。

print(site['sitename'], site['county'], site['aqi'], site['status'])： 逐行列印每個站點的名稱（sitename）、縣市（county）、空氣品質指標（aqi）和狀態（status）。這些鍵對應於 JSON 資料中每個站點的屬性。

```python
import json
with open('aqx_p_488.json',encoding='utf-8') as file:
    data:dict = json.loads(file.read())
    data_list:list[dict] = data["records"]

for site in data_list:
    print(site['sitename'],site['county'],site['aqi'],site['status'])
```
