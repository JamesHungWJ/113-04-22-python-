## json檔是一種複雜的資料結構
- json.loads 的作用是將 JSON 格式的字符串轉換成 Python 的數據結構
- list：列表



```python
import json
with open('aqx_p_488.json',encoding='utf-8') as file: #抓取aqx_p_488.json這個檔案
    data:dict = json.loads(file.read()) #json有提供一個function叫loads，將整個文字，轉成data
    data_list:list = data["records"]  
    #在aqx_p_488.json這個檔案裡查到records這個key，裡面放的是list。

for site in data_list:  #可以用for in迴圈將list裡面的一個一個抓出來。
    print(site)
```
