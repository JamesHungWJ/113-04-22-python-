## dictionary示意圖
- 純值得表示法是用    {key:    value, key: value} -->用的是大括號
- 下圖純值的表示用法  {"tw":"Taiwan","jp":"Japan"}






- csv：簡單的表格資料
- json：複雜的資料結構
- 以上均為純文字檔

```python
from csv import DictReader #從csv當中只import "Dictreader"這個功能而已，其他的都不要

with open('aqx_p_488.csv',encoding='utf-8',newline='') as file:  #使用csv，官方說明書中建議，最好新增一個newline，是一個空字串。
    reader:DictReader = DictReader(file)
    sites:list = list(reader)

for site in sites: #sites是一個list，所以可以for...in...一筆一筆抓
    print(type(site))  #dict=dictionary
    #print(site) 變成一個巢狀結構
    print(site['sitename'],site['county'],site['aqi'],site['status'][:5])
```
