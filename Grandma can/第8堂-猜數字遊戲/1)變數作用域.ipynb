## 全域變數、區域變數
- 全域變數：主程式定義「全域」的名稱空間，在主程式定義的變數是「全域變數」。
- 區域變數：個別函式定義「區域」的名稱空間，個別函式內定義的變數就是「區域變數」。



```python
e=10 #此即是全域變數
def abc (a,b): #此即是區域變數
    c=a+b
    print(c)

def efg():
    c = 10
    print(e)
#以上兩個def當中的變數c，是不一樣的，是完全獨立的，
#例如汽車，abc跟efg，雖然都是豐田汽車，但兩輛豐田汽車各自屬於不同人。

#在函式的內部才有區域變數，如果再if、while、for等迴圈中作用，就不是區域變數。
```
